```python
def caught_speeding(speed,is_birthday):
    Ticket=''
    if is_birthday==False:
        if speed<=60:
            Ticket= 'No Ticket'
        elif speed>=61 and speed<=80:
            Ticket= 'Small Ticket'
        else:
            Ticket= 'Big Ticket'
    else:
        if speed<=65:
            Ticket= 'No Ticket'
        elif speed>=66 and speed<=85:
            Ticket= 'Small Ticket'
        else:
            Ticket= 'Big Ticket'
    return Ticket

caught_speeding(81,False)
```
